```python
import xml.etree.ElementTree as ET
```

```python

tree = ET.parse('../Data/SilverbackOpen2018.xml')
root = tree.getroot()
```

```python
class _BBboyResults:
    def __init__(self,bboy_name):
        self.name = bboy_name
        self.judge_to_votes_dict = {}
        
    def add_data(judge_name,physical,artistic,interpretive):
        if judge_name in judge_to_votes_dict:
            current_tracker = self.judge_to_votes_dict[judge_name]
            current_tracker.add_votes
        else: 
            vote_tracker = _JudgeVoteTracker()
            vote_tracker.physical_diff = physical
            vote_tracker.artistic_diff = artistic
            vote_tracker.interpretive_diff = interpretive
            self.judge_to_votes_dict[judge_name] = vote_tracker
        
class _JudgeVoteTracker:
    def __init__(self):
        self.physical_diff = []
        self.artistic_diff = []
        self.interpretive_diff = []
        self.net_diff = []
    def add(self,physical,artistic,interpretive):
        self.physical_diff.append(physical)
        self.artistic_diff.append(artistic)
        self.interpretive_diff.append(interpretive)
        self.net_diff.append(physical+artistic+interpretive)
        
   
    
        
    
```

```python
bboy_results_dict = {}
for battle in root:
    competitors = battle.find("Competitors")
    competitor_1 = competitors.find("Competitor_1").text
    competitor_2 = competitors.find("Competitor_2").text
    
    #If competitors arent in already, create a Judge Vote Tracker for them
    if competitor_1 not in bboy_results_dict:
        bboy_results_dict[competitor_1] = _JudgeVoteTracker()
    if competitor_2 not in bboy_results_dict:
        bboy_results_dict[competitor_2] = _JudgeVoteTracker()
    judge_list = []
    judges = battle.find("Judges")
    judge_list.append(judges.find("Judge_1").text)
    judge_list.append(judges.find("Judge_2").text)
    judge_list.append(judges.find("Judge_3").text)
    judge_list.append(judges.find("Judge_4").text)
    judge_list.append(judges.find("Judge_5").text)
    judge_list.append(judges.find("Judge_6").text)
    judge_list.append(judges.find("Judge_7").text)
    
    round_list = battle.findall("Round")
    for round in round_list:
        bboy_1_vote_tracker = bboy_results_dict[competitor_1]
        bboy_2_vote_tracker = bboy_results_dict[competitor_2]

        physical_1 = float(round.find("Judge_1_Scores").find("Physical").text)
        artistic_1 = float(round.find("Judge_1_Scores").find("Artistic").text)
        interpretive_1 = float(round.find("Judge_1_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_1,artistic_1,interpretive_1)
        bboy_2_vote_tracker.add(-physical_1,-artistic_1,-interpretive_1)
        
        physical_2 = float(round.find("Judge_2_Scores").find("Physical").text)
        artistic_2 = float(round.find("Judge_2_Scores").find("Artistic").text)
        interpretive_2 = float(round.find("Judge_2_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_2,artistic_2,interpretive_2)
        bboy_2_vote_tracker.add(-physical_2,-artistic_2,-interpretive_2)
        
        physical_3 = float(round.find("Judge_3_Scores").find("Physical").text)
        artistic_3 = float(round.find("Judge_3_Scores").find("Artistic").text)
        interpretive_3 = float(round.find("Judge_3_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_3,artistic_3,interpretive_3)
        bboy_2_vote_tracker.add(-physical_3,-artistic_3,-interpretive_3)
        
        physical_4 = float(round.find("Judge_4_Scores").find("Physical").text)
        artistic_4 = float(round.find("Judge_4_Scores").find("Artistic").text)
        interpretive_4 = float(round.find("Judge_4_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_4,artistic_4,interpretive_4)
        bboy_2_vote_tracker.add(-physical_4,-artistic_4,-interpretive_4)
        
        physical_5 = float(round.find("Judge_5_Scores").find("Physical").text)
        artistic_5 = float(round.find("Judge_5_Scores").find("Artistic").text)
        interpretive_5 = float(round.find("Judge_5_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_5,artistic_5,interpretive_5)
        bboy_2_vote_tracker.add(-physical_5,-artistic_5,-interpretive_5)
        
        physical_6 = float(round.find("Judge_6_Scores").find("Physical").text)
        artistic_6 = float(round.find("Judge_6_Scores").find("Artistic").text)
        interpretive_6 = float(round.find("Judge_6_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_6,artistic_6,interpretive_6)
        bboy_2_vote_tracker.add(-physical_6,-artistic_6,-interpretive_6)
        
        physical_7 = float(round.find("Judge_7_Scores").find("Physical").text)
        artistic_7 = float(round.find("Judge_7_Scores").find("Artistic").text)
        interpretive_7 = float(round.find("Judge_7_Scores").find("Interpretive").text)
        bboy_1_vote_tracker.add(physical_7,artistic_7,interpretive_7)
        bboy_2_vote_tracker.add(-physical_7,-artistic_7,-interpretive_7)
        

    
```
